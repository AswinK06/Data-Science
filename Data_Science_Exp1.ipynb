```python
import pandas as pd

file_path = "/content/ipl_dataset(new).csv"
ipl_df = pd.read_csv(file_path)


print("First few rows of the dataset:")
print(ipl_df.head())


print("\nDataset Info:")
print(ipl_df.info())

print("\nSummary Statistics:")
print(ipl_df.describe(include="all"))

print("\nMissing Values:")
print(ipl_df.isnull().sum())

ipl_df[['Winning_Margin_Value', 'Winning_Margin_Type']] = ipl_df['Winning_Margin'].str.extract(r'(\d+)\s+(\w+)', expand=True)
ipl_df['Winning_Margin_Value'] = pd.to_numeric(ipl_df['Winning_Margin_Value'], errors='coerce')

ipl_df['Winning_Margin_Value'] = ipl_df['Winning_Margin_Value'].fillna(ipl_df['Winning_Margin_Value'].median())
ipl_df['Winning_Margin_Type'] = ipl_df['Winning_Margin_Type'].fillna("Unknown")

categorical_columns = ['Team_A', 'Team_B', 'Venue', 'Toss_Winner', 'Toss_Decision', 'Winner', 'Player_of_the_Match', 'Winning_Margin_Type']
ipl_df_encoded = pd.get_dummies(ipl_df, columns=categorical_columns, drop_first=True)

Q1 = ipl_df_encoded['Winning_Margin_Value'].quantile(0.25)
Q3 = ipl_df_encoded['Winning_Margin_Value'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

ipl_df_cleaned = ipl_df_encoded[(ipl_df_encoded['Winning_Margin_Value'] >= lower_bound) & (ipl_df_encoded['Winning_Margin_Value'] <= upper_bound)]

print("\nCleaned Dataset (First 5 Rows):")
print(ipl_df_cleaned.head())


cleaned_file_path = "cleaned_ipl_dataset.csv"
ipl_df_cleaned.to_csv(cleaned_file_path, index=False)
print(f"\nCleaned dataset saved as '{cleaned_file_path}'")
```
